### Преобразование столбцов
* map
* apply
* get_dummies - бинарное кодирование категориального признака (столбца)
* поэлементные операции (+, *, >, >= и т. д.)

```python
df["EncodedSex"] = df["Sex"].map({"male":-1, "female":1, "unknown":0})
```

```python
df["Sex"].head()
```

```python
df["EncodedSex"].head()
```

```python
def fun(elem):
    if elem < 30:
        return 1
    else:
        return 0
df["Young"] = df["Age"].apply(fun)
```

```python
df["Young"].head()
```

```python
df["Age"].head()
```

```python
df["Young"] = df["Age"].apply(lambda elem: 1 if elem > 30 else 0)
```

Сделать столбец Surname: первая часть имени (до ;)
Braund; Mr. Owen Harris -> Braund

```python
df["Surname"] = df["Name"].apply(lambda s: s[:s.find(";")])
```
